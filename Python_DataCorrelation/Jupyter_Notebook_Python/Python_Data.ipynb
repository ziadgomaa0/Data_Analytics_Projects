```python
#Create Correct year column
df_filtered['Real_year'] = df_filtered['released'].astype(str).str[9:13]
```

```python
df_filtered.sort_values(by= ['gross'] ,inplace = False , ascending = False)
```

```python
sns.regplot(x="gross", y="budget", data=df_filtered , scatter_kws = {"color" :"red"}, line_kws={"color":"blue"})
# plotting budget vs gross using seaborn
```

```python
# Correlation Matrix between all numeric columns

df_filtered.corr(method ='pearson')
```

```python
df_filtered.corr(method ='kendall')
```

```python
df_filtered.corr(method ='spearman')
```

```python
correlation_matrix = df_filtered.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize to assigns a random numeric value for each unique categorical value

df_filtered.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df_filtered.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
